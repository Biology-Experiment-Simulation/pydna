## Pydna Gel Electrophoresis Simulation

```python
import pydna
```

```python
from pydna.gel import weight_standard_sample
```

```python
st = weight_standard_sample('1kb+_GeneRuler')
```

```python
pUC19 = pydna.genbank("M77789")
```

```python
from Bio.Restriction import ZraI, BamHI
```

```python
frags = pUC19.cut(ZraI, BamHI)
```

```python
frags
```

```python
pydna.Gel( [ frags, st] ).run()
```
